### Outlier

```python
df['Year']=df['Year'].astype(int)
x=df['price'].astype(float)
df['price']=x.astype(int)
df['usdprice']=df['usdprice'].astype(float)*100
df['usdprice']=df['usdprice'].astype(int)
```

```python
df.head()
```

```python
df['price'].describe()
```

```python
plt.figure(figsize=(10,6))

plt.subplot(2,2,1)
sns.histplot(x=df['price'],kde=True)
plt.subplot(2,2,2)
sns.boxplot(x=df['price'])

plt.subplot(2,2,3)
sns.histplot(x=df['usdprice'],kde=True)
plt.subplot(2,2,4)
sns.boxplot(x=df['usdprice'])

plt.show()
```

```python
df['price'].describe()
```

```python
df.shape
```

```python
# Removing Outliers using Five number summary in price

colum=['price','usdprice']
for i in colum:
    percentile25=df[i].quantile(0.25)
    percentile75=df[i].quantile(0.75)
    IQR=percentile75 - percentile25
    min=percentile25 - 1.5*IQR
    max=percentile75 + 1.5*IQR
    #trimming
    df_new=df[(df[i]>min) & (df[i]<max)]
```

```python
plt.figure(figsize=(10,6))
plt.subplot(2,2,1)
sns.histplot(x=df_new['price'],kde=True)
plt.subplot(2,2,2)
sns.boxplot(x=df_new['price'])

plt.subplot(2,2,3)
sns.histplot(x=df_new['usdprice'],kde=True)
plt.subplot(2,2,4)
sns.boxplot(x=df_new['usdprice'])
```
