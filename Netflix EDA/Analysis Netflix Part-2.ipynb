```python
# Importing Libraries 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import plotly.express as px
import plotly.offline as pyo
sns.set_style("whitegrid")
```

```python
df=pd.read_csv('Netflix_analysis-2.csv')
df.drop(columns=['Unnamed: 0'],axis=1,inplace=True)
```

```python
df.head()
```

```python
sns.pairplot(data=df)
```

# 1. Analyzing Subscription ?

##### Basic ---> 0
##### Premium ---> 1
##### Standard ---> 2

```python
plt.figure(figsize=(25,20))
plt.subplot(2,2,1)
sns.countplot(x='Join Month',hue='Subscription Type',data=df)
plt.subplot(2,2,2)
sns.countplot(x='Join Year',hue='Subscription Type',data=df)
```

### As previously analyze from June to Oct Month many subscription as made but right data is showing major subscription are done in year 2022.

# 2. Analyzing country the netflix earn more revenue ?  

```python
# Monthly 
plt.figure(figsize=(20,15))
plt.subplot(2,2,1)
sns.countplot(x='Monthly Revenue',hue='Country',data=df)
plt.subplot(2,2,2)
sns.jointplot(x='Monthly Revenue',y='Subscription Type',hue='Country',data=df)
```

### US,Spain and Canada has most of the subscription with good revenue.

```python
fig=px.choropleth(df, locations="Country", locationmode='country names', color="Monthly Revenue",
              color_continuous_scale='Viridis')
pyo.plot(fig, filename = 'Monthly-Revenue-Subscription-Type-Country.html')
```

# 3. Which age has done maximum subscription in Year 2021, 2022 and 2023? 

```python
plt.figure(figsize=(20,15))
plt.subplot(2,2,1)
df.groupby('Age').sum()['Subscription Type'].plot(kind='barh',color='red')
plt.subplot(2,2,2)
sns.countplot(x='Age',hue='Join Year',data=df)
```
